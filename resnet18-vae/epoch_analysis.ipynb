```python
import pickle
import torch

class VaeUnpickler(pickle.Unpickler):
    def __init__(self, file):
         super().__init__(file)
    def persistent_load(self, pid):
        print(pid)
        return torch.FloatStorage()
        # return super().persistent_load(pid)

with open('data.pkl', 'rb') as f:
    upkl = VaeUnpickler(file=f)
    upkl.load()

print(upkl)
```

```python
with open('../20240327_epoch_14.zip', 'rb') as f:
    model = torch.load(f)
    
msd = model['model_state_dict']
total = 0
for layer in msd:
    print(f"{layer}: {msd[layer].dtype} {msd[layer].shape} ({msd[layer].numel() * 4} bytes)") # size of int
    total += msd[layer].numel() * 4

print(f"total = {total}")
```
