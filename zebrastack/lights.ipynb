```python
user_id = "kLTFiYXQW8TPIyiKSphdR1MoOephHb2YrPreQUtZ"
response = requests.get(f"http://10.0.0.227/api/{user_id}/lights")
assert response.status_code == 200
```

```python
import json
lights = json.loads(response.content)
len(lights)
```

```python
import pandas as pd
def light_properties(light):
    properties = {'name':light['name']}
    properties.update(light['config'])    
    properties.pop('startup')    
    properties.update(light['capabilities']['control'])
    properties.pop('colorgamut')
    properties.update(light['state'])
    properties.pop('effect')
    properties.pop('mode')
    return properties

pd.DataFrame([light_properties(lights[key]) for key in lights],
             lights.keys())
```
